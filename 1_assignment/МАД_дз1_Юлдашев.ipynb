### Задача 1 [1 балл].
Пробег (в 1000км) автомобиля "Лада Калина" до капитального ремонта двигателя является нормальной случайной величиной с неизвестным математическим ожиданием μ и известной дисперсией 49. По выборке из 20 автомобилей постройте доверительный интервал для математического ожидания пробега с уровнем доверия 0.95.

```python
X = [321.90 , 307.26, 194.18, 373.78, 287.03, 251.33, 249.19, 303.62,
       333.47, 264.23, 277.13, 325.44, 227.72, 299.75, 267.21, 316.72,
       269.45, 260.69, 328.81, 335.34]

import numpy as np
from scipy import stats

data = np.array(X)
n = len(data)

sample_mean = np.mean(data) # Среднее
sample_std = np.std(data) # Стандартное отклонение
confidence_level = 0.95 # Уровень доверия
df = n - 1 # Степени свободы

t_value = stats.t.ppf(1 - (confidence_level / 2), df) # т-критерий для заданного уровня доверия
margin_of_error = t_value * (sample_std / np.sqrt(n)) # Предел погрешностей

lower_bound = sample_mean - margin_of_error # Нижняя граница доверительного интервала
upper_bound = sample_mean + margin_of_error # Верхняя граница доверительного интервала

print("Доверительный интервал для математического ожидания пробега:")
print(f"Нижняя граница: {lower_bound:.2f} км")
print(f"Верхняя граница: {upper_bound:.2f} км")
```

### Задача 2 [2 балла].
Бутерброд подбросили 200 раз, 105 раз он упал маслом вниз и 95 маслом вверх. Проверьте гипотезу о том, что бутерброд с равной вероятностью падает маслом вверх и вниз.

```python
import numpy as np
from scipy import stats

n = 200
successes = 105
failures = 95

alpha = 0.05 # Уровень значимости

# Нулевая гипотеза (H0): Вероятность падения маслом вверх = Вероятность падения маслом вниз
# Альтернативная гипотеза (H1): Вероятность падения маслом вверх ≠ Вероятность падения маслом вниз

res = stats.binomtest(successes, n, p=0.5) # Встроенная функция тестирования нулевой гипотезы

if res.pvalue <= alpha:
  print("H0 отвергается. Бутерброд не падает с равной вероятностью маслом вверх и вниз.")
else:
  print("H0 не отвергается. Бутерброд, возможно, падает с равной вероятностью маслом вверх и вниз.")
```

Закон будет принят, если ЗА проголосует более половины депутатов. Какое минимальное количество депутатов нужно опросить на выходе с заседания, чтобы построить критерий, проверяющий гипотезу о том, что закон не принят, против альтернативы (информации от агента), с уровнем значимости 5% и мощностью не менее 80%?

```python
from scipy.stats import binom_test
import numpy as np
import math

def number_of_respondents(p0, p1, N, alpha, beta):
  beta = 1- beta
  z_alpha = stats.norm.ppf(1 - (1 - alpha_))
  z_beta = stats.norm.ppf(1 - (1 - beta))

  n = ((math.sqrt(p1*(1 - p1)) * z_beta + math.sqrt(p0*(1 - p0)) * z_alpha)/(p1 - p0)) ** 2
  return int(np.ceil(n))

N = 450
p0 = 0.5
p1 = 0.7
alpha_ = 0.05
beta = 0.8
print("Количество необходимых респондентов для опроса: ", number_of_respondents(p0, p1, N, alpha, beta))
```
